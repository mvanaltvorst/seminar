# Commodity Price Data
This notebook was used while exploring CMO-Historical-Data-Monthly.xlsx

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read Data

```python
# world bank data on commodity prices
df_cmo1 = pd.read_excel(
    "./../../assets/CMO-Historical-Data-Monthly.xlsx",
    sheet_name="Monthly Prices",
    header=6
)
# Set first column as index
df_cmo1.set_index('Unnamed: 0', inplace=True, drop=True)
```

```python
df_cmo1
```

```python
# world bank data on commodity prices
df_cmo2 = pd.read_excel(
    "./../../assets/CMO-Historical-Data-Monthly.xlsx",
    sheet_name="Monthly Indices",
    header=9
)
# Set first column as index
df_cmo2.set_index('Unnamed: 0', inplace=True, drop=True)
```

```python
df_cmo2
```

```python
df_cmo = pd.concat([df_cmo1, df_cmo2], axis=1)
```

```python
df_cmo = df_cmo[['CRUDE_PETRO', 'iNATGAS', 'iAGRICULTURE', 'iMETMIN', 'iPRECIOUSMET']]
```

```python
# Replace all non numeric values with NaN
df_cmo = df_cmo.apply(pd.to_numeric, errors='coerce')
df_cmo = df_cmo.dropna()
df_cmo = df_cmo.rename_axis('Date')
```

```python
# Make the dataframe quarterly
df_cmo.index = pd.to_datetime(df_cmo.index, format='%YM%m')
df_cmo = df_cmo.resample('Q').mean()
```

```python
df_cmo
```

```python
df_cmo_change = df_cmo.pct_change()
```

```python
df_cmo_change
```

## Visualise

```python
df_cmo_change.corr()
```

```python
df_cmo_change.plot()
```

```python
df_cmo.plot()
```

```python
correlation_matrix = df_cmo_change.corr()

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(correlation_matrix)

# Sort the eigenvalues and eigenvectors by decreasing eigenvalues
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]

# Calculate the percentage of total variance each PC accounts for
variance_explained = sorted_eigenvalues / sum(sorted_eigenvalues) * 100

# Plotting the percentage of variance explained by each principal component
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.pie(variance_explained, labels=[f'PC{i+1}' for i in range(len(variance_explained))],
        autopct='%1.1f%%', startangle=140)
plt.title('Variance Explained by Each Principal Component')

# Plotting the eigenvectors
plt.subplot(1, 2, 2)
for i in range(len(sorted_eigenvectors)):
    plt.plot(sorted_eigenvectors[:, i], marker='o', label=f'PC{i+1}')
plt.title('Eigenvectors (Principal Components)')
plt.xlabel('Index of Eigenvector Element')
plt.xticks(range(len(sorted_eigenvalues)))
plt.ylabel('Component Weight')
plt.legend()

plt.tight_layout()
plt.show()
```
