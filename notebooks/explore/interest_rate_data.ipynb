# Interest rate data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_interest_rate = pd.read_excel('./../../assets/Monthly_interest_rates.xlsx')
df_interest_rate.drop(columns=['DATAFLOW_ID:Dataflow ID', 'KEY:Timeseries Key', 'FREQ:Frequency', 'Unit', 'Unit multiplier',
       'TIME_PERIOD:Period'], inplace=True)
```

```python
df_interest_rate.set_index('REF_AREA:Reference area', inplace=True)
```

```python
df_interest_rate = df_interest_rate.T
```

```python
# rename the index to date
df_interest_rate.index.rename('Date', inplace=True)
df_interest_rate.index = pd.to_datetime(df_interest_rate.index)
# Coerce non numeric to NaN
```

```python
df_interest_rate
```

```python
# drop columns with all NaN values
df_interest_rate = df_interest_rate.dropna(axis=1, how='all')
```

```python
# Count how many countries have data for each date
df_interest_rate['count'] = df_interest_rate.count(axis=1)
```
