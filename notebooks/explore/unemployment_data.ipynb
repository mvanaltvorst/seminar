# Unemployment Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read Data

```python
df_unemployment = pd.read_csv("./../../assets/unemployment-ilo.csv")
df_unemployment.drop(['Sex', 'Age'], axis=1, inplace=True)
df_unemployment.set_index('Quarter', inplace=True)
```

```python
# Drop columns with all NaN values
df_unemployment.dropna(axis=1, how='all', inplace=True)
```

```python
df_unemployment.index = pd.PeriodIndex(df_unemployment.index, freq='Q').to_timestamp()
```

```python
df_unemployment.sort_index(inplace=True)
```

```python
countries = df_unemployment.columns
```

```python
# Plot the median
df_unemployment['Median'] = df_unemployment.median(axis=1)
```

```python
df_unemployment['Median'].plot()
```

```python
# Count how many countries have data for each quarter
df_unemployment['Count'] = df_unemployment[countries].count(axis=1)
```

```python
df_unemployment['Count'].plot()
```
